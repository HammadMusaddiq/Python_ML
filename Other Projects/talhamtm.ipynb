```python
#Prediction models (NB, RF, SVM, DT)
models = []
models.append(('NB', GaussianNB()))
models.append(('RF', RandomForestClassifier()))
models.append(('SVM', SVC(gamma='auto')))
models.append(('DT', DecisionTreeClassifier()))

results = []
resultsavg = []
names = []
resultsstd = []

for name, model in models:
    cv_results = cross_val_score(model, xm, y, cv=kfold, scoring='accuracy')
    
    results.append(cv_results)
    resultsavg.append(cv_results.mean())
    resultsstd.append(cv_results.std())
    names.append(name)
    print('%s: %f (%f)' % (name, cv_results.mean(), cv_results.std()))
```

```python
#ANN model (MLP Classifier sk-learn)
model_ANN = MLPClassifier(alpha=0.05, max_iter=300, hidden_layer_sizes=(15,), random_state=1) #15 Neurons with 1 hidden layer

cv_results_ANN = cross_val_score(model_ANN, xm, y, cv=kfold, scoring='accuracy')
cv_results_ANN_mean = cv_results_ANN.mean()
cv_results_ANN_std = cv_results_ANN.std()
cv_results_ANN_name = 'ANN'
print('%s: %f (%f)' % (cv_results_ANN_name, cv_results_ANN_mean, cv_results_ANN_std))
```

```python
#Keras Deep Neural Network with sk-learn (Keras Classifier) + (Sequentional Model)
def baseline_model():
    # create model
    model_DNN = Sequential()
    model_DNN.add(Dense(12, input_dim=11, activation='relu'))
    model_DNN.add(Dense(8, activation='relu'))
    model_DNN.add(Dense(8, activation='relu'))
    model_DNN.add(Dense(2, activation='softmax'))
    # Compile model
    model_DNN.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model_DNN

model1_DNN = KerasClassifier(build_fn=baseline_model, epochs=200, batch_size=5, verbose=0)

cv_results_DNN = cross_val_score(model1_DNN, xm, y, cv=kfold, scoring='accuracy')
cv_results_DNN_mean = cv_results_DNN.mean()
cv_results_DNN_std = cv_results_DNN.std()
cv_results_DNN_name = 'DNN'
print('%s: %f (%f)' % (cv_results_DNN_name, cv_results_DNN_mean, cv_results_DNN_std))
```
