```python
import numpy as np
import pandas as pd
import re
```

```python
#importing dataset
data = pd.read_json("IMRAN_MENTAL_HEALTH3031.json")
```

```python
data.head(3)
```

```python
data.info()
```

```python
#Null values in the data
data.isnull().values.any()
```

```python
#Empty cells in the data
empty = np.where(data.applymap(lambda x: x == ''))
print(empty)
```

```python
#count empty cells
len(empty[0])
```

```python
#Empty cells in location column
#Count value show that all empty cells are of location column because of same count value
cou = data.loc[data['location'] == '']
cou
```

```python
len(cou)
```

```python
#Replacing empty cells of location value to 'Unknown'
data['location'] = data["location"].replace({"": "Unknown"})
data.head(5)
```

```python
#Location empty cells renamed to Unknown
len(data.loc[data['location'] == 'Unknown'])
```

```python
#Created new column of 'data' from 'created_at'
data['date'] = pd.DatetimeIndex(data['created_at']).date
data.head(3)
```

```python
data["date"].value_counts().head(10)
```

```python
len(data["date"].value_counts())
```

```python
#As we can see in the 'date' column there is only change in dates. So only considering dates
data['date'] = data['created_at'].dt.day
data.head(3)
```

```python
#Created new column of 'time' from 'created_at'
data['time'] = pd.DatetimeIndex(data['created_at']).time
data.head(3)
```

```python
data["time"].value_counts().head(10)
```

```python
len(data["time"].value_counts())
```

```python
#Converting 'time' column data to only hours
data['time'] = data['created_at'].dt.hour
data.head(3)
```

```python
#Data bining of 'created_at' column in times
bins = [-1, 6, 12, 18, 24]
labels = ['0-6', '7-12', '13-18', '19-23']
data['time_bins'] = pd.cut(data['time'], bins=bins, labels=labels)
data.head(3)
```

```python
#Extracting 'tags' from text column
tags_ = data['text'].astype(str).str.findall(r'@.*?(?=\s|$)')
data['tags_in_text'] = tags_.astype(str).str.replace("[","").str.replace("]","").str.replace(":","").str.replace("'","")
data.head(3)
```

```python
#Extracting 'hashtags' from text column
hashtags_ = data['text'].astype(str).str.findall(r'#.*?(?=\s|$)')
data['hashtags_in_text'] = hashtags_.astype(str).str.replace("[","").str.replace("]","").str.replace(":","").str.replace("'"," ")
data.head(3)
```

```python
#Exported new generated data to CSV
data.to_csv('Tweets_Health_Data.csv')  
```
