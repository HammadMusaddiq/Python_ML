### cofidance interval

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import random
import math
%matplotlib inline
```

```python
df=pd.read_csv("FinalcleanedData.csv")
```

```python
df
```

```python
df=df.dropna()
```

```python
size_mapping = {'poor': 0,
                'Middle': 1,
                'upper': 2}
```

```python
df['Family_status'] = df['Family_status'].map(size_mapping)
```

```python
df
```

```python
point_estimates_dict = {}
Confidence_interval_dict = {}
```

```python
def Point_Estimates_Confidence_interval(column,qvalue):
    if df[column].dtype==np.int64:
        no_of_yes=np.sum(df[column]==1)
        print("No of Yes in",column,no_of_yes)
        point_Estimate=no_of_yes/len(df)
    else:
        point_Estimate=df[column].mean()
    
    print('Point Estimate of',column,":",point_Estimate)
    name='Point Estimate of '+column
    point_estimates_dict[name] = point_Estimate

    sample_mean=point_Estimate
    z_critical = stats.norm.ppf(q = qvalue)  # Get the z-critical value*

    print("z-critical value:",z_critical)              # Check the z-critical value
    if df[column].dtype==np.int64:
        pop_stdev =math.sqrt( point_Estimate*((len(df)-no_of_yes)/len(df) )) # Get the population standard deviation
    else:
        pop_stdev =df[column].std() # Get the population standard deviation                    
    
    print("Standard deviation of",column,":",pop_stdev) 
    margin_of_error = z_critical * (pop_stdev/math.sqrt(len(df)))

    confidence_interval = (sample_mean - margin_of_error,
                           sample_mean + margin_of_error)  

    print("Confidence interval of ",column,":",confidence_interval)
    name2='Confidence interval of '+column
    Confidence_interval_dict[name2] = confidence_interval
    print("  ")
    print("  ")
```

```python
list=df.columns
for i in list:
    Point_Estimates_Confidence_interval(i,0.95)
```
