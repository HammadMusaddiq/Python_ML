```python
%%time
import datetime 
start_time = '00:01:00'
end_time = '00:07:00' 

start = datetime.datetime.strptime(start_time, '%H:%M:%S')
end = datetime.datetime.strptime(end_time, '%H:%M:%S')

time_zero = datetime.datetime.strptime('00:00:00', '%H:%M:%S')
complete_time = str(end - start)

import subprocess

# (subprocess.run) is fast
# According to the python documentation subprocess.run waits for the process to end
#!ffmpeg -i Dubai.mp4 -ss 00:01:20 -to 00:05:00 -c:v copy -c:a copy output2.mp4 -y
#subprocess.run(["ffmpeg",'-i','Dubai.mp4','-ss',start_time,'-to',complete_time,'-c:v','copy','-c:a','copy','output3.mp4','-y'])

cmd = ["ffmpeg",'-i','Dubai.mp4','-ss',start_time,'-to',complete_time,'-c:v','copy','-c:a','copy','output33.mp4','-y']
process = subprocess.Popen(cmd,stdout=subprocess.PIPE,stderr=subprocess.PIPE)
process.wait()
    
#!ffmpeg -i Dubai.mp4 -ss 00:01:00 -t complete_time -c:v copy -c:a copy output1.mp4 -y
```
